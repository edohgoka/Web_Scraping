```python
from bs4 import BeautifulSoup
import requests
```

Now we need to give the **URL**, **request parameters** and **soup** the element we want to extract.

```python
url = "https://www.w3schools.com/python/ref_func_next.asp"
```

```python
req = requests.get(url)
```

```python
soup = BeautifulSoup(req.text, "html.parser")
```

```python
print(soup.title.text)
```

**Find all links on a web page**

```python
for link in soup.find_all("a"):
    print(link.get("href"))
```

**Geting paragraphs and Headings**

```python
paragraphs = soup.find_all("p")
for paragraph in paragraphs:
    print(paragraph.text)
```

```python
## We have different types of heading: h1, h2, h3, h4, ...
headings = soup.find_all("h2")
headings
```

Exctraction **Images**, **Styles** and **Div tags**

```python
## The <div> tag defines a division or a section in an HTML document
div_texts = soup.find_all("div")
div_texts[1]
```

```python
styles = soup.find_all("style")
print(styles[0])
```

```python
images = soup.find_all("img")
images
```

```python
## Let's show the text that are in bold
bold_texts = soup.find_all("b")
bold_texts
```

```python
## Let's show the italic texts
italic_texts = soup.find_all("i")
italic_texts
```

**Prettify**

```python
## To get in a pretty way the source code of a whole web page
pretf = soup.prettify()
print(pretf)
```

**Export to a csv file**

```python
import csv
```

```python
with open("links.csv", "w") as f:
    csv_writer = csv.writer(f)
    for link in soup.find_all("a"):
        l = link.get("href")
        # Now add the link
        csv_writer.writerow([l])
```
